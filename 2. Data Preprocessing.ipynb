# Read dataset from PostgreSQL

```python
import pandas as pd
from sqlalchemy import create_engine

# Connection string (sesuaikan dengan yang Anda gunakan)
engine = create_engine('postgresql+psycopg2://postgres:0000@localhost:5434/Employee_Retention_Project')

# Query untuk membaca data
query = "SELECT * FROM employee_data"

# Membaca data ke dalam DataFrame
with engine.begin() as connection:
    df = pd.read_sql(query, connection)

df
```

# Check Data 

## Check for null values, unique values, and data types

```python
data_summary = pd.DataFrame({
    'Total Null': df.isna().sum(),
    'Percentage Null': df.isna().sum() * 100 / len(df),
    'Number of Unique Values': df.nunique(),
    'Data Type': df.dtypes,
    'Unique Values': df.apply(lambda col: col.dropna().unique().tolist()),
    'Mode': df.apply(lambda col: col.mode().iloc[0] if not col.mode().empty else None)
})

# Format Percentage Null
data_summary['Percentage Null'] = data_summary['Percentage Null'].apply(
    lambda x: f"{x:.2f}%" if x != 0 else "0%"
)

data_summary
```

## Filling null values
Fill with the median (to avoid being affected by data distribution) for numerical features.

```python
numerical_has_null = [  ]

# Convert columns to numeric (handling errors and nulls)
df[numerical_has_null] = df[numerical_has_null].apply(pd.to_numeric, errors='coerce')

# Fill null values with the median
df[numerical_has_null] = df[numerical_has_null].fillna(df[numerical_has_null].median())
```

The column 'IkutProgramLOP' is mostly empty. However, we are keeping it for now as it might be useful for answering questions like, "Among the employees who joined the LOP program, what percentage of them resigned?" For machine learning analysis, however, this column will be dropped.

The column 'AlasanResign' contains empty values, which may indicate that the employee has not resigned. These empty values can be replaced with 'masih_bekerja', which is also the mode of the data.

```python
df['AlasanResign'] = df['AlasanResign'].fillna('masih_bekerja')
```

# Fix the values in Column 'PernahBekerja'

the 'PernahBekerja' column has 2 unique values, '1' and 'yes', which is actually the same values.

```python
df['PernahBekerja'] = df['PernahBekerja'].replace({'yes': '1'})
```

# Drop some unusable columns

```python
data_summary = pd.DataFrame({
    'Total Null': df.isna().sum(),
    'Percentage Null': df.isna().sum() * 100 / len(df),
    'Number of Unique Values': df.nunique(),
    'Data Type': df.dtypes,
    'Unique Values': df.apply(lambda col: col.dropna().unique().tolist())
})

# Format Percentage Null
data_summary['Percentage Null'] = data_summary['Percentage Null'].apply(
    lambda x: f"{x:.2f}%" if x != 0 else "0%"
)

data_summary
```

Since 'PernahBekerja' only have 1 value, let's drop the column

```python
df = df.drop(columns='PernahBekerja')
```

# Change to proper datatypes

```python
# string
string_cols = ['Username','EnterpriseID','StatusPernikahan','JenisKelamin','StatusKepegawaian','Pekerjaan', 'JenjangKarir', 
               'PerformancePegawai','AsalDaerah', 'HiringPlatform','NomorHP','Email', 'TingkatPendidikan','AlasanResign']

df[string_cols] = df[string_cols].astype(str)

# numerical (integer)
num_cols = ['SkorSurveyEngagement','SkorKepuasanPegawai','JumlahKeikutsertaanProjek','JumlahKeterlambatanSebulanTerakhir', 'JumlahKetidakhadiran']
df[num_cols] = df[num_cols].astype(int)

# date
date_cols = ['TanggalLahir', 'TanggalHiring', 'TanggalPenilaianKaryawan', 'TanggalResign']
df[date_cols] = df[date_cols].apply(pd.to_datetime, errors='coerce')
```

# Adjust the 'None' format

```python
import numpy as np
df = df.where(pd.notna(df), None)
df = df.replace({'-':None,'None': None, np.nan: None})
df
```

```python
data_summary = pd.DataFrame({
    'Total Null': df.isna().sum(),
    'Percentage Null': df.isna().sum() * 100 / len(df),
    'Number of Unique Values': df.nunique(),
    'Data Type': df.dtypes,
    'Unique Values': df.apply(lambda col: col.dropna().unique().tolist()),
    'Mode': df.apply(lambda col: col.mode().iloc[0] if not col.mode().empty else None)
})

# Format Percentage Null
data_summary['Percentage Null'] = data_summary['Percentage Null'].apply(
    lambda x: f"{x:.2f}%" if x != 0 else "0%"
)

data_summary
```

# Write Preprocessed data to PostgreSQL

```python
from tqdm import tqdm
engine = create_engine('postgresql+psycopg2://postgres:0000@localhost:5434/Employee_Retention_Project')
batch_size = 1000
total_rows = len(df)
with engine.begin() as connection:
    for i in tqdm(range(0, total_rows, batch_size), desc="Inserting rows", dynamic_ncols=True):
        df.iloc[i:i + batch_size].to_sql('employee_data_preprocessed', connection, if_exists='replace', index=False)
```
